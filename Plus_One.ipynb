## Solution 1: While loop

```python
    def plusOne(digits):
      index = len(digits) - 1

      while index >= 0 and digits[index] == 9:
          digits[index] = 0
          index -= 1

      if index < 0:
          digits.insert(0, 1)
      else:
          digits[index] += 1

      print(digits)

    plusOne([9])
```
