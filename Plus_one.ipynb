## 66. Plus one
You are given a large integer represented as an integer array digits, where each digits[i] is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's.

Increment the large integer by one and return the resulting array of digits.



Example 1:

Input: digits = [1,2,3]
Output: [1,2,4]
Explanation: The array represents the integer 123.
Incrementing by one gives 123 + 1 = 124.
Thus, the result should be [1,2,4].
Example 2:

Input: digits = [4,3,2,1]
Output: [4,3,2,2]
Explanation: The array represents the integer 4321.
Incrementing by one gives 4321 + 1 = 4322.
Thus, the result should be [4,3,2,2].
Example 3:

Input: digits = [9]
Output: [1,0]
Explanation: The array represents the integer 9.
Incrementing by one gives 9 + 1 = 10.
Thus, the result should be [1,0].

## Solution 1: Carry propagation using while loop

```python
def plusOne(digits):
  index = len(digits) - 1

  while index >= 0 and digits[index] == 9:
    digits[index] = 0
    index -= 1

  if index < 0:
    digits.insert(0, 1)
  else:
    digits[index] += 1

  print(digits)

plusOne([1, 2, 3, 9])
```

## Solution 2: Carry propagation using for loop

```python
def plusOne(digits):
  for i in range(len(digits)-1, -1, -1):
    if digits[i] < 9:
      digits[i] += 1
      break
    elif digits[i] == 9 and i == 0:
      digits[i] = 0
      digits.insert(0, 1)
    else:
      digits[i] = 0
      continue
  print(digits)

plusOne([1, 9, 9])
```

## Solution 3: String conversion approach

```python
def plusOne(digits):
    num = int(''.join(map(str, digits))) + 1
    print([int(digit) for digit in str(num)])

plusOne([9, 9])
```

## Solution 4: Using library(reduce)

```python
from functools import reduce

def plusOne(digits):
    num = reduce(lambda acc, d: acc*10 + d, digits) + 1
    print([int(x) for x in str(num)])
plusOne([1, 2, 3])
```

## Solution 5: Stack style

```python
def plusOne(digits):
    carry = 1

    for i in range(len(digits)-1, -1, -1):
        digits[i] += carry
        carry = digits[i] // 10
        digits[i] %= 10

    if carry:
        digits.insert(0, carry)

    print(digits)

plusOne([1, 2, 9, 9])
```
